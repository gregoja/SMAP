```python
import pandas as pd

df = pd.read_json('data/test_data/thumbnails.json')

print(df)
```

```python
import pandas as pd
import os
import urllib.request

folder_path = 'data/test_data/downloaded/'
os.makedirs(folder_path, exist_ok=True)

for index, row in df.iterrows():
    image_url = str(row).split('http://')[-1].split()[0]
    image_name = f"image_{index}.jpg"

    try:
        full_image_url = f'http://{image_url}'
        urllib.request.urlretrieve(full_image_url, os.path.join(folder_path, image_name))
        print(f"Downloaded: {image_name}")
    except Exception as e:
        print(f"Failed to download {image_name}: {e}")
```
